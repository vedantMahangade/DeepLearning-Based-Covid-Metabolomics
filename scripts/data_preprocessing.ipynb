## Data Cleaning
Since both datasets are obtained from different studies, there is inconsitancy in the data.
- The metadata files contain infomation regarding the subject involved ie Subject/PatientID, Group/Condition, Sex, Age.
- The metabolite file contains expression data for each metabolite in each subject.


### Renaming Metabolites
The inconsistancy in the datasets is predominant in metabolite names.<br>
For example in TM171 a metabolite is named C0, which also has a record in PM108, but with a different name Carnitine (synonyms for C0)
- There are various databases available online, which could be used to obtain these synonyms.
- The one used here is the [The Human Metabolome Database (hmdb)](https://hmdb.ca/).

```python
# Create a dictionary to replace metabolite names in TM161 dataset to match those in PM108
replace_metabolite_dict = {
    'Methionine-sulfoxide':'Methionine sulfoxide',
    'SM(OH) C14:': 'SM(d18:0/14:1(9Z)(OH))',
    'SM C16:0': 'SM(d18:0/16:0)',
    'SM(OH) C16:1': 'SM(d18:0/16:1(9Z)(OH))',
    'SM C18:1': 'SM(d18:1/18:1(9Z))',
    'SM C18:0': 'SM(d18:0/18:0)',
    'SM C20:2': 'SM(d18:0/20:2(11Z,14Z))',
    'SM(OH) C24:1': 'SM(d18:0/24:1(15Z)(OH))',
    'C0': 'Carnitine',
    'C2': 'Acetylcarnitine',
    'C3': 'Propionylcarnitine',
    'C4': 'Butyrylcarnitine',
    'C5': 'Valerylcarnitine',
    'C4OH': 'Hydroxybutyrylcarnitine',
    'C8': 'Octanoylcarnitine',
    'C10:2': 'Decadienoylcarnitine',
    'C10:1': 'Decenoylcarnitine',
    'C10': 'Decanoylcarnitine',
    'C12:1': 'Dodecenoylcarnitine',
    'C12': 'Dodecanoylcarnitine',
    'C14:1': 'Tetradecenoylcarnitine',
    'C16': 'Hexadecanoylcarnitine',
    'C18:2': 'Octadecadienylcarnitine',
    'C18:1': 'Octadecenoylcarnitine'
}
```

```python
print(pm108_metabolite.columns)
print(tm161_metabolite.columns)
```

```python
tm161_metabolite = tm161_metabolite.rename(columns=replace_metabolite_dict)
```

### Renaming patient classes
In case of metadata, both datasets have different classifications for patient groups.<br>
PM108 contains a group called CONTROL, for which corresponding name in TM161 is G1.<br>
For ease of further analysis and to have a good class distribution, we will create 3 classes:
- Control
- Covid
- Recovered

```python
print("Patient classes in PM108:", set(pm108_metadata["Group"]))
print("Patient classes in TM161:", set(tm161_metadata["Group"]))
```

```python
replace_group_dict = {
    'COVID-19': "Covid", 
    'CONTROL': "Control", 
    'RECOVERED': "Recovered", 
    'CLASS A': "Recovered", 
    'CLASS B': "Recovered", 
    'G1': "Control", 
    'G2': "Covid", 
    'G3': "Covid", 
    'G4': "Covid"
}
```

```python
pm108_metadata = pm108_metadata.replace({"Group": replace_group_dict})
tm161_metadata = tm161_metadata.replace({"Group": replace_group_dict})
```

```python
print("Patient classes in PM108:", set(pm108_metadata["Group"]))
print("Patient classes in TM161:", set(tm161_metadata["Group"]))
```

## Data Cleaning
Now that our datasets are merged, we can clean the data to prepare for EDA and further analysis

### Check for duplicates

```python
print("Number of duplicate rows:", data.duplicated().sum())
```

```python
data = data.drop_duplicates()
```

### Make sure all Metabolite columns are numeric

```python
missing_values = data.isnull().sum()
print("\nMissing Values per Column:\n", missing_values[missing_values > 0])
```

```python
data.select_dtypes(include="object",exclude="float64").dtypes
```

```python
columns_to_convert = [
    'Putrescine', 'Serotonin', 'Trimethylamine N-oxide', 'p-Hydroxyhippuric acid',
    'LysoPC a C26:1', 'LysoPC a C28:0', 'PC aa C36:6', 'Hydroxybutyrylcarnitine',
    'Decadienoylcarnitine', 'Decenoylcarnitine', 'Dodecenoylcarnitine', 'Dodecanoylcarnitine',
    'Octadecadienylcarnitine'
]
```

```python
# Check for non-numeric values and handle them
for col in columns_to_convert:
    # Check for rows where the column has non-numeric values
    non_numeric = data[col].apply(pd.to_numeric, errors='coerce').isnull()
    if non_numeric.any():
        print(f"\nNon-numeric values found in column '{col}':")
        print(data[col][non_numeric].head())

    # Convert to float (non-numeric values will be turned into NaN)
    data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
data.select_dtypes(include="object",exclude="float64").dtypes
```

### Check for missing values

```python
missing_values = data.isnull().sum()
print("\nMissing Values per Column:\n", missing_values[missing_values > 0])
```

#### Handle missing or empty values
- Numeric column: fill with median
- Categorical columns: fill with mode

```python
numerical_cols = data.select_dtypes(include=[np.number]).columns
categorical_cols = data.select_dtypes(include=[object, 'category']).columns
```

```python
# Impute numerical columns with mean
imputer_num = SimpleImputer(strategy='mean')
data[numerical_cols] = imputer_num.fit_transform(data[numerical_cols])
```

```python
# Impute categorical columns with mode (for categorical variables like Group, Sex)
imputer_cat = SimpleImputer(strategy='most_frequent')
data[categorical_cols] = imputer_cat.fit_transform(data[categorical_cols])
```

### Saving the final dataset

```python
data.to_csv("../data/covid19_metabolites.csv")
```
